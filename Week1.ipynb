## 2. 向量

### 2.1 创建向量

```r
# 8. 使用c函数创建向量
numvec <- c(1, 2, 3, 4, 5)
numvec
class(numvec)
length(numvec)

charvec <- c('Hello', 'World', '!')
charvec
class(charvec)
length(charvec)

logvec <- c(T, T, FALSE)
logvec
class(logvec)
length(logvec)
```

```r
# 9. 冒号运算符
a <- 1:9
a
class(a)
```

```r
# 10. seq函数：产生一个数列的向量
seq(-5, 5)
seq(-5, 5, by=.3)
seq(-5, 5, length=7)
seq(from=-5, length=7, by=.3)
```

```r
# 11. req函数：重复元素
x <- 1:4
rep(x, 3)  # times=3
rep(x, each=3)
```

### 2.2 类型转换

```r
# 12. 隐式类型转换
x <- c(1.7, 'abc')  # 数字转成文本
x
y <- c(TRUE, 2)  # 逻辑转成数字
y
z <- c('abc', T)  # 逻辑转成文本
z
```

```r
# 13. 强制类型转换
x <- 0:6
as.logical(x)
as.character(x)

as.numeric(c('a', 'b', 'c'))  # 转换失败时触发警告，得到NA值
as.logical(c('a', 'b', 'c'))
```

### 2.3 向量运算

```r
# 14. 向量运算即逐元素运算
x <- 50:100
any(x > 80)
any(x < 80)
all(x > 80)

x <- 1:10
y <- 21:30
x + y
x / y

sum(x)
mean(x)
var(x)
sd(x)
```

```r
# 15. 重复利用的逐元素运算（循环补齐）
1:10 + 2
1:2 + 1:10
1:3 + 1:10  # 重复后无法匹配长度触发警告，但仍能运算
```

### 2.4 索引与子集

```r
# 16. 使用整数下标
(1:10)[2]
(1:10)[2:3]  # 使用冒号向量索引
(1:10)[0]
(1:10)[-2]
(1:10)['a']
(1:10)[100]
```

```r
# 17. 使用整数向量
c('a', 'b', 'c', 'd')[c(1, 3, 3, 3)]
c('a', 'b', 'c', 'd')[-(1:3)]  # 排除元素
c('a', 'b', 'c', 'd')[rep(c(1, 2, 2, 1), times=4)]
```

```r
# 18. 使用逻辑向量
c('a', 'b', 'c', 'd')[c(F, T, T, F)]
c('a', 'b', 'c', 'd')[c(F, T)]  # 循环补齐
c('a', 'b', 'c', 'd')[c(F, T, T, T, T, F)]  # 超出部分中有TRUE的，取NA
```

```r
# 19. 使用比较运算符
x <- 1:10
x[x %% 2 == 0]

x <- c(1, 5, 3)
x[x > 2 | x == 1]
x[x > 2 & x < 4]

x <- c(5, 2, 3, 8)
x[x > 4] <- 0
x
```

```r
# 20. 使用subset函数
x <- 1:5
subset(x, x > 2)
```

```r
# 21. 使用which函数
x <- c(8, 1, 5, 2)
which(x > 2)
which.min(x)
```

```r
# 22. 使用元素的名称
fruit <- c(5, 10, 1, 20)
names(fruit) <- c('orange', 'banana', 'apple', 'peach')
fruit[c('apple', 'orange')]
```

### 2.5 空向量与缺失值

```r
# 23. 直接使用类型函数创建空向量
e <- numeric()
e
e <- character(0)  # length=0
e
```

```r
# 24. 缺失值
x <- c(NA, 1, NA)
x
x[7] <- NA
x
length(x)
length(x) <- 10
x
```

```r
# 25. 缺失值判断
v <- c(1:4, NA, 3, NA)

is.na(v)
v[v > 2]
subset(v, v > 2)
```

### 2.6 随机数

```r
# 26. 随机正态分布的向量
rnorm(10)
x <- rnorm(10, 20, 2)  # 生成10个均值20，标准差2的正态分布样本
x
```

```r
# 延伸
dnorm(10, 20, 2)  # 概率密度函数：均值20，标准差2的正态分布在x=10时的概率密度
pnorm(10, 20, 2)  # 累积分布函数：均值20，标准差2的正态分布在x=10时的累积概率
qnorm(0.975, 20, 2)  # 分位数函数：均值20，标准差2的正态分布累积概率为0.975对应的分位数
```

```r
# 27. 汇总统计量
summary(x)
```

```r
# 28. 随机种子
set.seed(1)
rnorm(5)
rnorm(5)
set.seed(1)
rnorm(5)  # 种子相同时，产生的随机数相同
```

```r
# 29. 抽样
sample(1:12, 9)
sample(1:12, 9, replace=TRUE)
sample(1:12)  # 重新排列
```
