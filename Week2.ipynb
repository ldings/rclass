## 1. R数据结构

### 1.1 数组和矩阵

```r
# 1. 向量与数组、矩阵的关系
z <- 1:1500  # 一维

dim(z) <- c(3, 5, 100)  # 设置维度
class(z)

dim(z) <- c(15, 100)  # 二维
class(z)
```

```r
# 2. 使用array函数创建数组
a <- array(1:20, dim=2)  # 一维数组，循环补齐（只取两个元素）
a

a <- array(1:20, dim=c(2))  # 一维数组
a

class(a)
```

```r
#
a <- array(1:20, dim=c(4, 5))  # 二维数组
a

a <- array(1:20, c(4, 6))  # 二维数组
class(a)

a <- array(1, c(3, 2))  # 二维数组
a
```

```r
# 注意：载入数组的方法是按列顺序，这与FORTRAN语言相同，不同于C语言
x <- array(1:24, dim=c(3, 4, 2))
x
```

```r
# 3. 数组的索引
x[3, 2, 1]
# x[1, 2, 3]  # 第三个维度超出
x[3,, 1]
x[3,,]
```

```r
# 4. 使用数组索引数组
x <- array(1:20, c(4, 5))
x

i <- array(c(1:3, 3:1), c(3, 2))
i

x[i] <- 0
x
```

```r
# 5. 使用matrix函数创建矩阵
m <- matrix(1:6, nrow=2)
m

matrix(1:6, ncol=3)

matrix(1:6, 2, byrow=TRUE)  # nrow=2，指定按行排序
```

```r
# 注意：若行列元素不一致时，进行截取或循环补齐，并触发警告
matrix(1:6, 2, 4)
matrix(1:6, 2, 2)
```

```r
# 6. 使用比较运算符索引矩阵
m <- matrix(1:6, 2)
m

m[1,] <= 3

m[, m[1,] <= 3]
```

```r
# 7. 使用which函数索引矩阵
m <- matrix(1:6, 2)
which(m > 2)
```

```r
# 8. 使用元素名称索引矩阵
m <- matrix(1:6, 2)
rownames(m) <- c('row1', 'row2')
colnames(m) <- c('col1', 'col2', 'col3')
m

m['row2', 'col3']
m[2, 3]  # 仍可用下标索引
```

```r
# 9. 替换值
m <- matrix(1:6, 2)
m[1,] <- 999
m
```

```r
# 10. 矩阵的横向合并与纵向合并
m <- matrix(1:6, 2)
p <- matrix(7:12, 2)

cbind(m, p)  # 要求行数一致
rbind(m, p)  # 要求列数一致
```

```r
# 11. 按行/列加和与平均值
m <- matrix(1:12, 3, 4)
m

rowSums(m)
rowMeans(m)

colSums(m)
colMeans(m)

sum(m)
mean(m)
```

```r
# 12. 矩阵逐元素算术运算
A = matrix(1:6, 2)
A * A  # 逐元素乘法
#A %*% A  # 矩阵乘法要求矩阵形状满足要求

B = matrix(1:6, 3)
A %*% B
```

```r
# 13. apply函数：将函数作用到数组的若干维度上
x <- matrix(1:20, 10)
x

apply(x, 2, mean)  # margin=2，作用的维度（作用到列上）

colMeans(x)
```

```r
#
x <- matrix(1:6, 3)
f <- function(x) { return (x / c(2, 8)) }
apply(x, 1, f)  # 作用到行上
```

```r
# 14. t函数：转置矩阵
x = matrix(1:6, 3)
x
t(x)
```

```r
# 27. 创建因子
x <- factor(c('yes', 'yes', 'no', 'yes', 'no'))
x

table(x)  # 生成频数表
```

```r
# 28. 有序因子
temp <- factor(c('high', 'low', 'high', 'medium', 'low'), levels=c('low', 'medium', 'high'), ordered=TRUE)
temp

min(temp)  # 有序因子可计算最值
```

```r
# 29. tapply函数按照因子分组聚合
temp <- c(10, 4, 27, 31, -10, -25, 35)
tlevel <- factor(c('medium', 'medium', 'high', 'high', 'low', 'low', 'high'))

tapply(temp, tlevel, mean)
tapply(temp, tlevel, range)
```

```r
#
x <- data.frame(gender=factor(c('M', 'M', 'F', 'M', 'F', 'F')),
                ages=c(21, 25, 27, 54, 20, 32),
                income=c(24000, 31000, 44000, 31000, 22000, 60000))
x

x$over25 <- ifelse(x$age > 25, 1, 0)  #新增over25列
x

tapply(x$income, list(x$gender, x$over25), mean)  # 按照两列分组
```

```r
# 30. split函数分组
x
split(x, x$gender)
```

## 4. R语言编程结构

```r
# 36. If语句
clod <- FALSE
temp <- 3

if (temp < 10) {
    cold <- TRUE
}

cold
```

```r
#
x <- 0

if (x > 3) {
    y <- 1
} else if (x <= 3 && x > 0) {
    y <- 0
} else {
    y <- -1
}

y
```

```r
# 37. for循环语句
x <- c(1, 5, 4, 3, 2)

for (i in 1:length(x)) {
    print(x[i])
}
```

```r
#
x <- matrix(1:6, 2)
x

for (i in seq_len(nrow(x))) {
    for (j in seq_len(ncol(x))) {
        print(x[i, j])
    }
}
```

```r
# 38. while循环语句
i <- 1

while (i < 5) {
    print(i)
    i <- i + 1
}
```

```r
# 39. repeat循环语句
v <- c('Hello', 'loop')
cnt <- 2

repeat {
    print(v)
    cnt <- cnt + 1
    
    if (cnt > 5) {
        break
    }
}
```

```r
# 40. next语句
v <- LETTERS[1:6]

for (i in v) {
    if (i == "D") {
        next
    }
    print(i)
}
```

```r
# 41. 函数
f <- function(a, b) {
    y <- a * b
    return(y)
}

f(2, 3)
f(b=3, a=2)
```

```r
# 42. 变量作用域
w <- 12

f <- function(y) {
    d <- 8
    h <- function() {
        return(d * (w + y))
    }
    return(h())
}

f(2)
```

```r
#
y <- 10

f <- function(x) {
    y <- 2  # 局部变量
    y ^ 2 + g(x)
}

g <- function(x) {
    x * y  # 全局变量
}

f(3)
```
